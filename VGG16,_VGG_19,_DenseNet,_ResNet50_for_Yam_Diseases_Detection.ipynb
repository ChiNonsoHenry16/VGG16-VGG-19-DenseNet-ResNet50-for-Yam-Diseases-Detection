#VGG 16

```python
#  MOUNT GOOGLE DRIVE
from google.colab import drive
drive.mount('/content/drive', force_remount=True)  # Added force_remount=True

#  IMPORTS
import os
import numpy as np
import tensorflow as tf
from collections import Counter
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input   # CHANGED HERE
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import matplotlib.pyplot as plt
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve, auc
from sklearn.preprocessing import label_binarize
import seaborn as sns
import time

# DATA PATHS AND SETUP
train_path = "/content/drive/My Drive/Mbajidataset6000Images/train"
val_path = "/content/drive/My Drive/Mbajidataset6000Images/val"
test_path = "/content/drive/My Drive/Mbajidataset6000Images/test"

batch_size = 32
target_size = (224, 224)  # VGG16 default input size

print("\n===== Configuration =====")
print(f"Train data path: {train_path}")
print(f"Val data path:   {val_path}")
print(f"Test data path:  {test_path}")
print(f"Batch size:      {batch_size}")
print(f"Target img size: {target_size}")

# IMAGE DATA GENERATORS
# Use VGG16 preprocess_input (scales input pixels as required by that model)
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=50,
    width_shift_range=0.3,
    height_shift_range=0.3,
    shear_range=0.2,
    zoom_range=0.3,
    fill_mode='nearest',
    brightness_range=[0.6, 1.4],
    channel_shift_range=20.0,
    horizontal_flip=True,
    vertical_flip=True
)
test_val_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

train_generator = train_datagen.flow_from_directory(
    train_path,
    target_size=target_size,
    batch_size=batch_size,
    class_mode='sparse',
    shuffle=True,
    seed=42
)
val_generator = test_val_datagen.flow_from_directory(
    val_path,
    target_size=target_size,
    batch_size=batch_size,
    class_mode='sparse',
    shuffle=False
)
test_generator = test_val_datagen.flow_from_directory(
    test_path,
    target_size=target_size,
    batch_size=batch_size,
    class_mode='sparse',
    shuffle=False
)

num_classes = len(train_generator.class_indices)
class_names = list(train_generator.class_indices.keys())
print("\nClasses:", train_generator.class_indices)

counter = Counter(train_generator.classes)
print("Training set class counts:", dict(counter))

# CLASS WEIGHTS
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(train_generator.classes),
    y=train_generator.classes
)
class_weights = dict(zip(np.unique(train_generator.classes), class_weights))
print("Class Weights:", class_weights)

# PREPROCESS CHECK
np.random.seed(0)
sample_image = np.random.randint(0, 256, size=(1,224,224,3), dtype=np.uint8)  # Adjusted size
sample_preprocessed = preprocess_input(sample_image.astype(np.float32))
print(f"VGG16 preprocess_input, value range: {np.min(sample_preprocessed)} to {np.max(sample_preprocessed)}")
# VGG16 preprocess_input subtracts mean RGB, so outputs may be positive or negative

# BUILD MODEL
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(224,224,3))   # CHANGED HERE
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = BatchNormalization()(x)
x = Dense(384, activation='relu', kernel_regularizer=l2(2e-4))(x)
x = Dropout(0.7)(x)
x = Dense(128, activation='relu', kernel_regularizer=l2(1e-4))(x)
x = Dropout(0.5)(x)
predictions = Dense(num_classes, activation='softmax')(x)
model = Model(inputs=base_model.input, outputs=predictions)
for layer in base_model.layers:
    layer.trainable = False

optimizer = Adam(learning_rate=1e-4)
model.compile(optimizer=optimizer, loss="sparse_categorical_crossentropy", metrics=["accuracy"])
model.summary()

#  TRAIN
early_stop = EarlyStopping(monitor='val_loss', patience=7, verbose=1, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=4, verbose=1, min_lr=1e-6)
start_time = time.time()
history = model.fit(
    train_generator,
    epochs=25,
    validation_data=val_generator,
    callbacks=[early_stop, reduce_lr],
    class_weight=class_weights,
    verbose=1
)
train_time = time.time() - start_time
print(f"\nTraining time: {train_time/60:.2f} minutes")

#  FINE-TUNE
for layer in base_model.layers[-40:]:
    layer.trainable = True
optimizer_finetune = Adam(learning_rate=4e-5)
model.compile(optimizer=optimizer_finetune, loss="sparse_categorical_crossentropy", metrics=["accuracy"])
history_finetune = model.fit(
    train_generator,
    epochs=12,
    validation_data=val_generator,
    callbacks=[early_stop, reduce_lr],
    class_weight=class_weights,
    verbose=1
)
train_time_fine = time.time() - start_time
print(f"Total time (fine-tune included): {train_time_fine/60:.2f} min")

def merge_histories(history1, history2):
    merged = {}
    for k in history1.history.keys():
        merged[k] = list(history1.history[k]) + list(history2.history[k])
    return merged

full_history = merge_histories(history, history_finetune)
epochs = range(1, len(full_history['loss']) + 1)

#  EVALUATION
train_loss, train_acc = model.evaluate(train_generator, verbose=0)
val_loss, val_acc = model.evaluate(val_generator, verbose=0)
test_loss, test_acc = model.evaluate(test_generator, verbose=0)
print(f"\nFINAL METRICS")
print(f"Train   Loss: {train_loss:.4f} | Accuracy: {train_acc:.4f}")
print(f"Val     Loss: {val_loss:.4f} | Accuracy: {val_acc:.4f}")
print(f"Test    Loss: {test_loss:.4f} | Accuracy: {test_acc:.4f}")

# PLOT LEARNING CURVES
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(epochs, full_history['loss'], label='Training Loss')
plt.plot(epochs, full_history['val_loss'], label='Validation Loss')
plt.axhline(test_loss, color='r', linestyle='--', label='Test Loss')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss Curve')

plt.subplot(1,2,2)
plt.plot(epochs, full_history['accuracy'], label='Training Accuracy')
plt.plot(epochs, full_history['val_accuracy'], label='Validation Accuracy')
plt.axhline(test_acc, color='r', linestyle='--', label='Test Accuracy')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy Curve')
plt.tight_layout()
plt.show()

# CONFUSION MATRIX
test_generator.reset()
y_true = test_generator.classes
y_pred_prob = model.predict(test_generator)
y_pred = np.argmax(y_pred_prob, axis=1)
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()

print(classification_report(y_true, y_pred, target_names=class_names))

# ROC-AUC (per class, macro, weighted)
from sklearn.preprocessing import label_binarize

y_true_bin = label_binarize(y_true, classes=np.arange(num_classes))
roc_auc_per_class = []

plt.figure(figsize=(10,6))
for i in range(num_classes):
    try:
        fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_pred_prob[:, i])
        roc_auc = auc(fpr, tpr)
        roc_auc_per_class.append(roc_auc)
        plt.plot(fpr, tpr, lw=2, label=f'{class_names[i]} (AUC = {roc_auc:.2f})')
    except ValueError:
        roc_auc_per_class.append(np.nan)

plt.plot([0, 1], [0, 1], 'k--', label='Random')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Each Class')
plt.legend()
plt.tight_layout()
plt.show()

# Per-class ROC-AUC BAR PLOT
plt.figure(figsize=(10, 4))
plt.bar(range(num_classes), roc_auc_per_class, color='teal')
plt.ylim(0, 1.05)
plt.xticks(range(num_classes), class_names, rotation=30)
plt.title("Per-class ROC-AUC scores on test set")
plt.ylabel("ROC-AUC")
for i, v in enumerate(roc_auc_per_class):
    plt.text(i, v + 0.01, f"{v:.2f}" if not np.isnan(v) else "NA", ha='center')
plt.tight_layout()
plt.show()

# Macro and weighted ROC-AUC scores
mean_macro_auc = np.nanmean(roc_auc_per_class)
try:
    weighted_auc = roc_auc_score(y_true_bin, y_pred_prob, average='weighted', multi_class='ovr')
except ValueError:
    weighted_auc = np.nan

print(f"Test Macro ROC-AUC:    {mean_macro_auc:.4f}")
print(f"Test Weighted ROC-AUC: {weighted_auc if not np.isnan(weighted_auc) else 'Could not compute'}")

print("\nPipeline Complete!")
```

```python
# MOUNT GOOGLE DRIVE
from google.colab import drive
drive.mount('/content/drive', force_remount=True)  # Added force_remount=True

# IMPORTS
import os
import numpy as np
import tensorflow as tf
from collections import Counter
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.densenet import DenseNet121, preprocess_input   # CHANGED HERE
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import matplotlib.pyplot as plt
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve, auc
from sklearn.preprocessing import label_binarize
import seaborn as sns
import time

# DATA PATHS AND SETUP
train_path = "/content/drive/My Drive/Mbajidataset6000Images/train"
val_path = "/content/drive/My Drive/Mbajidataset6000Images/val"
test_path = "/content/drive/My Drive/Mbajidataset6000Images/test"

batch_size = 32
target_size = (224, 224)  # DenseNet121 input size matches VGG19

print("\n===== Configuration =====")
print(f"Train data path: {train_path}")
print(f"Val data path:   {val_path}")
print(f"Test data path:  {test_path}")
print(f"Batch size:      {batch_size}")
print(f"Target img size: {target_size}")

# IMAGE DATA GENERATORS
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=50,
    width_shift_range=0.3,
    height_shift_range=0.3,
    shear_range=0.2,
    zoom_range=0.3,
    fill_mode='nearest',
    brightness_range=[0.6, 1.4],
    channel_shift_range=20.0,
    horizontal_flip=True,
    vertical_flip=True
)
test_val_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

train_generator = train_datagen.flow_from_directory(
    train_path,
    target_size=target_size,
    batch_size=batch_size,
    class_mode='sparse',
    shuffle=True,
    seed=42
)
val_generator = test_val_datagen.flow_from_directory(
    val_path,
    target_size=target_size,
    batch_size=batch_size,
    class_mode='sparse',
    shuffle=False
)
test_generator = test_val_datagen.flow_from_directory(
    test_path,
    target_size=target_size,
    batch_size=batch_size,
    class_mode='sparse',
    shuffle=False
)

num_classes = len(train_generator.class_indices)
class_names = list(train_generator.class_indices.keys())
print("\nClasses:", train_generator.class_indices)

counter = Counter(train_generator.classes)
print("Training set class counts:", dict(counter))

#  CLASS WEIGHTS
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(train_generator.classes),
    y=train_generator.classes
)
class_weights = dict(zip(np.unique(train_generator.classes), class_weights))
print("Class Weights:", class_weights)

#  PREPROCESS CHECK
np.random.seed(0)
sample_image = np.random.randint(0, 256, size=(1,224,224,3), dtype=np.uint8)  # Adjusted size
sample_preprocessed = preprocess_input(sample_image.astype(np.float32))
print(f"DenseNet121 preprocess_input, value range: {np.min(sample_preprocessed)} to {np.max(sample_preprocessed)}")
# DenseNet121 preprocess_input subtracts mean RGB channels; output can be positive or negative.

# BUILD MODEL
base_model = DenseNet121(weights='imagenet', include_top=False, input_shape=(224,224,3))  # CHANGED HERE
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = BatchNormalization()(x)
x = Dense(384, activation='relu', kernel_regularizer=l2(2e-4))(x)
x = Dropout(0.7)(x)
x = Dense(128, activation='relu', kernel_regularizer=l2(1e-4))(x)
x = Dropout(0.5)(x)
predictions = Dense(num_classes, activation='softmax')(x)
model = Model(inputs=base_model.input, outputs=predictions)
for layer in base_model.layers:
    layer.trainable = False

optimizer = Adam(learning_rate=1e-4)
model.compile(optimizer=optimizer, loss="sparse_categorical_crossentropy", metrics=["accuracy"])
model.summary()

# TRAIN
early_stop = EarlyStopping(monitor='val_loss', patience=7, verbose=1, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=4, verbose=1, min_lr=1e-6)
start_time = time.time()
history = model.fit(
    train_generator,
    epochs=25,
    validation_data=val_generator,
    callbacks=[early_stop, reduce_lr],
    class_weight=class_weights,
    verbose=1
)
train_time = time.time() - start_time
print(f"\nTraining time: {train_time/60:.2f} minutes")

# FINE-TUNE
for layer in base_model.layers[-40:]:
    layer.trainable = True
optimizer_finetune = Adam(learning_rate=4e-5)
model.compile(optimizer=optimizer_finetune, loss="sparse_categorical_crossentropy", metrics=["accuracy"])
history_finetune = model.fit(
    train_generator,
    epochs=12,
    validation_data=val_generator,
    callbacks=[early_stop, reduce_lr],
    class_weight=class_weights,
    verbose=1
)
train_time_fine = time.time() - start_time
print(f"Total time (fine-tune included): {train_time_fine/60:.2f} min")

def merge_histories(history1, history2):
    merged = {}
    for k in history1.history.keys():
        merged[k] = list(history1.history[k]) + list(history2.history[k])
    return merged

full_history = merge_histories(history, history_finetune)
epochs = range(1, len(full_history['loss']) + 1)

# EVALUATION
train_loss, train_acc = model.evaluate(train_generator, verbose=0)
val_loss, val_acc = model.evaluate(val_generator, verbose=0)
test_loss, test_acc = model.evaluate(test_generator, verbose=0)
print(f"\nFINAL METRICS")
print(f"Train   Loss: {train_loss:.4f} | Accuracy: {train_acc:.4f}")
print(f"Val     Loss: {val_loss:.4f} | Accuracy: {val_acc:.4f}")
print(f"Test    Loss: {test_loss:.4f} | Accuracy: {test_acc:.4f}")

# PLOT LEARNING CURVES
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(epochs, full_history['loss'], label='Training Loss')
plt.plot(epochs, full_history['val_loss'], label='Validation Loss')
plt.axhline(test_loss, color='r', linestyle='--', label='Test Loss')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss Curve')

plt.subplot(1,2,2)
plt.plot(epochs, full_history['accuracy'], label='Training Accuracy')
plt.plot(epochs, full_history['val_accuracy'], label='Validation Accuracy')
plt.axhline(test_acc, color='r', linestyle='--', label='Test Accuracy')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy Curve')
plt.tight_layout()
plt.show()

# CONFUSION MATRIX
test_generator.reset()
y_true = test_generator.classes
y_pred_prob = model.predict(test_generator)
y_pred = np.argmax(y_pred_prob, axis=1)
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()

print(classification_report(y_true, y_pred, target_names=class_names))

# ROC-AUC (per class, macro, weighted)
from sklearn.preprocessing import label_binarize

y_true_bin = label_binarize(y_true, classes=np.arange(num_classes))
roc_auc_per_class = []

plt.figure(figsize=(10,6))
roc_auc_per_class = []  # Make sure this list is initialized before this block

for i in range(num_classes):
    try:
        fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_pred_prob[:, i])
        roc_auc = auc(fpr, tpr)
        roc_auc_per_class.append(roc_auc)
        plt.plot(fpr, tpr, lw=2, label=f'{class_names[i]} (AUC = {roc_auc:.2f})')
    except ValueError:
        roc_auc_per_class.append(np.nan)

plt.plot([0, 1], [0, 1], 'k--', label='Random')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Each Class')
plt.legend()
plt.tight_layout()
plt.show()

# Per-class ROC-AUC BAR PLOT
plt.figure(figsize=(10, 4))
plt.bar(range(num_classes), roc_auc_per_class, color='teal')
plt.ylim(0, 1.05)
plt.xticks(range(num_classes), class_names, rotation=30)
plt.title("Per-class ROC-AUC scores on test set")
plt.ylabel("ROC-AUC")
for i, v in enumerate(roc_auc_per_class):
    plt.text(i, v + 0.01, f"{v:.2f}" if not np.isnan(v) else "NA", ha='center')
plt.tight_layout()
plt.show()

# Macro and weighted ROC-AUC scores
mean_macro_auc = np.nanmean(roc_auc_per_class)
try:
    weighted_auc = roc_auc_score(y_true_bin, y_pred_prob, average='weighted', multi_class='ovr')
except ValueError:
    weighted_auc = np.nan

print(f"Test Macro ROC-AUC:    {mean_macro_auc:.4f}")
print(f"Test Weighted ROC-AUC: {weighted_auc if not np.isnan(weighted_auc) else 'Could not compute'}")

print("\nPipeline Complete!")
```
